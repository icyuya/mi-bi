```python
import cv2
import os

# 入力動画のパス
# video_path = 'video/NAICe1_WIN_20250519_09_25_56_Pro.mp4'
# video_path = 'video/NAICe2_WIN_20250519_09_29_21_Pro.mp4'
video_path = 'video/NAICe5_WIN_20250519_09_39_43_Pro.mp4'

# 静止画の出力先フォルダ
output_dir = 'image/'
os.makedirs(output_dir, exist_ok=True)

# 動画ファイルを読み込み
cap = cv2.VideoCapture(video_path)

# フレームレート（fps）を取得
fps = cap.get(cv2.CAP_PROP_FPS)

# 総フレーム数
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

# 動画の長さ（秒）
duration_sec = total_frames / fps
print(f"動画時間: {duration_sec:.2f} 秒")

# 1秒ごとのフレーム位置で静止画を保存
for sec in range(int(duration_sec)):
    # 保存したいフレーム番号
    frame_number = int(sec * fps)
    cap.set(cv2.CAP_PROP_POS_FRAMES, frame_number)  # フレーム位置を設定

    ret, frame = cap.read()
    if ret:
        filename = os.path.join(output_dir, f"frame_{sec:04d}.jpg")
        cv2.imwrite(filename, frame)
        print(f"保存: {filename}")
    else:
        print(f"読み込み失敗: {sec} 秒のフレーム")

cap.release()

```

```python
import os
import shutil
import math

# 元の画像フォルダ（2161枚）
input_dir = 'image/'

# 出力フォルダ（300枚）
output_dir = 'train/'
os.makedirs(output_dir, exist_ok=True)

# 画像ファイル一覧（並び順を保証）
all_images = sorted([f for f in os.listdir(input_dir) if f.endswith('.jpg')])
total_images = len(all_images)

# 等間隔のステップ数を計算
step = total_images / 300

selected_indices = [math.floor(i * step) for i in range(300)]
selected_images = [all_images[i] for i in selected_indices]

# 選ばれた画像をコピー
for filename in selected_images:
    src_path = os.path.join(input_dir, filename)
    dst_path = os.path.join(output_dir, filename)
    shutil.copyfile(src_path, dst_path)

print(f"{len(selected_images)} 枚の画像を等間隔に抽出して {output_dir} に保存しました。")
```

```python
# ライブラリのインポート
import numpy as np
import cv2
from matplotlib import pyplot as plt

# カラー画像の読み込み
# 第2引数：グレースケール指定
gray_org_img = cv2.imread('image/frame_0000.jpg',0)


# Cannyアルゴリズム実行
# 第2引数：ヒステリアスしきい値処理の最小しきい値
# 第3引数：ヒステリアスしきい値処理の最大しきい値
canny_edges = cv2.Canny(gray_org_img,200,200)


# 元画像描画
plt.subplot(121),plt.imshow(gray_org_img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])

# エッジ検出後の画像描画
plt.subplot(122),plt.imshow(canny_edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
import cv2
import numpy as np

def adjust_gamma(image, gamma=1.0):
    inv_gamma = 1.0 / gamma
    table = np.array([(i / 255.0) ** inv_gamma * 255 for i in range(256)]).astype("uint8")
    return cv2.LUT(image, table)

def unsharp_mask(image, kernel_size=(5, 5), alpha=1.5, beta=-0.5):
    blurred = cv2.GaussianBlur(image, kernel_size, 0)
    sharp = cv2.addWeighted(image, alpha, blurred, beta, 0)
    return sharp

# 画像読み込み
img = cv2.imread("image/frame_0000.jpg")

# グレースケール変換
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imwrite("aaaaaaa/step1_gray.jpg", gray)

# ノイズ除去（ガウシアンブラー）
blurred = cv2.GaussianBlur(gray, (5, 5), 0)
cv2.imwrite("aaaaaaa/step2_blurred.jpg", blurred)

# コントラスト強調（CLAHE）
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
enhanced = clahe.apply(blurred)
cv2.imwrite("aaaaaaa/step3_clahe.jpg", enhanced)

# === 新しい補正処理（ガンマ補正とシャープ化） ===
gamma_corrected = adjust_gamma(enhanced, gamma=1.5)
cv2.imwrite("aaaaaaa/step4_gamma.jpg", gamma_corrected)

sharpened = unsharp_mask(gamma_corrected)
cv2.imwrite("aaaaaaa/step5_sharpened.jpg", sharpened)

# エッジ検出（Canny）
edges = cv2.Canny(sharpened, 20, 60)
cv2.imwrite("aaaaaaa/step6_edges.jpg", edges)
```
